```python
from datetime import datetime
```

```python
headers = ['ID', 'Date of flight', 'Flight number', 'Departure point', 'Arrival point',
           'Scheduled Time departure', 'Scheduled Time arrival', 'Flight status', 'Aircraft Code',
           'target']
header = ['ID', 'Date of flight', 'Flight number', 'Departure point', 'Arrival point',
           'Scheduled Time departure', 'Scheduled Time arrival', 'Flight status', 'Aircraft Code']
```

```python
train = pd.read_csv('Train.csv', names=headers, low_memory=False)
test = pd.read_csv('Test.csv', names=header, low_memory=False)
```

```python
train = train.drop(0, axis=0)
test = test.drop(0, axis=0)
```

```python
train['Date of flight'] = pd.to_datetime(train['Date of flight'])
train['Scheduled Time departure'] = pd.to_datetime(train['Scheduled Time departure'])
train['Scheduled Time arrival'] = pd.to_datetime(train['Scheduled Time arrival'].apply(lambda x: x.replace('.', ':')))
train['month'] = train['Date of flight'].dt.month
train['year'] = train['Date of flight'].dt.year
```

```python
train['time of departure'] = train['Scheduled Time departure'].dt.time
train['time of arrival'] = train['Scheduled Time arrival'].dt.time
```
